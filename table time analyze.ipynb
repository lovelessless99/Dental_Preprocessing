```python
from datetime import datetime
import pandas as pd
import glob
```

```python
def subtract_time(subdir_time, table_time):
        print(subdir_time, table_time)
        subdir_time = subdir_time.split(" ")[1]
        table_time  = table_time.split(" ")[1]
        
        if subdir_time == table_time:
                return 0
        
        subdir_time = subdir_time[:4] + "20" + subdir_time[4:]
        table_time  = table_time[:4] + "20" + table_time[4:]

        sub_convert = datetime.strptime(subdir_time  , "%m%d%Y")
        table_convert = datetime.strptime(table_time, "%m%d%Y")
        
        return table_convert - sub_convert
        # print(sub_convert)
        # print(table_convert)
        # print(sub_convert-table_convert)
```

```python
sub_dir = [ i for i in glob.iglob("Data/*") ]

for i in sub_dir:
        items = [ i for i in glob.iglob("%s/*" % i) if not i.endswith(".json") ]
        
        if len(items) > 1:
                sub_dir, table = items
                
                if table.endswith("x"):
                        table, sub_dir = sub_dir, table
                
                
                sub_dir_time = sub_dir.split("\\")[2]
                table_time   = table.split("\\")[2]
                
                diff_time = subtract_time(sub_dir_time, table_time)
                if diff_time != 0 :
                        print(diff_time.days)
                else:
                        print(diff_time)

```
